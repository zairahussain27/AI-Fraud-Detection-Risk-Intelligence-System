```python
#TO CHECK THE COLUMN NAMES IN THE DATAFRAME AND RENAME THEM TO LOWERCASE AND REPLACE SPACES WITH UNDERSCORES

trans_df.columns=trans_df.columns.str.lower().str.replace(" ","_")
```

```python
#TO SAVE THE CLEANED DATAFRAME TO A NEW CSV FILE NAMED 'transactions_cleaned.csv' IN THE 'data' FOLDER

trans_df.to_csv('../data/transactions_cleaned.csv', index=False)
```
