# ***DATASET EXPLORATION***

```python
import pandas as pd
```

```python
#LOAD THE DATASET

trans_df = pd.read_csv('../data/transactions_raw.csv')
```

```python
#TO CHECK INFORMATION ABOUT THE DATAFRAME

trans_df.info()
```

```python
#TO CHECK THE FIRST 5 ROWS OF THE DATAFRAME
trans_df.head()
```

```python
#TO CHECK THE SHAPE (NO OF ROWS AND COLUMN) OF THE DATAFRAME

trans_df.shape
```

```python
#TO CHECK TOTAL NO OF NULL VALUES IN THE DATAFRAME

trans_df.isnull().sum()
```

```python
#TO COUNT VALUES IN THE FRAUD

trans_df['is_fraud'].value_counts()
```

```python
#TO CHECK THE PERCENTAGE OF FRAUD AND NON-FRAUD TRANSACTIONS

trans_df['is_fraud'].value_counts(normalize=True)*100
```

```python
#TO CHECK THE COLUMN NAMES IN THE DATAFRAME AND RENAME THEM TO LOWERCASE AND REPLACE SPACES WITH UNDERSCORES

trans_df.columns=trans_df.columns.str.lower().str.replace(" ","_")
```

```python
#TO CHECK THE DATA TYPES OF EACH COLUMN IN THE DATAFRAME
trans_df.dtypes
```

```python
#CONVERT TRANSACTION DATETIME COLUMN INTO PROPER DATETIME FORMAT

trans_df['trans_date_trans_time']=pd.to_datetime(trans_df['trans_date_trans_time'])
```

```python
#CREATE SEPTATE DATE COLUMN (USED FOR DAILY FRAUD TRENDS AND GROOUPING)

trans_df['trans_date']=trans_df['trans_date_trans_time'].dt.date

trans_df['trans_time']=trans_df['trans_date_trans_time'].dt.time
```

```python
#CONVERT DOB, TRANSACTION DATE TO PROPER DATETIME FORMAT

trans_df['dob']=pd.to_datetime(trans_df['dob'],errors='coerce')

trans_df['trans_date']=pd.to_datetime(trans_df['trans_date'],errors='coerce')
```

```python
#RENAME FIRST, LAST NAME AND TRANSACTION DATE TIME COLUMNS FOR BETTER CLARITY

trans_df=trans_df.rename(columns={'first':'firstname','last':'lastname'})

trans_df=trans_df.rename(columns={'trans_date_trans_time':'trans_datetime'})
```

```python
trans_df.describe()
```

```python
#TO SAVE THE CLEANED DATAFRAME TO A NEW CSV FILE NAMED 'transactions_cleaned.csv' IN THE 'data' FOLDER

trans_df.to_csv('../data/transactions_cleaned.csv', index=False)
```
